Background Layer Plotting: The background layer plotting typically involves creating the overall structure of the plot.                                             Here we use functions like :- plt.figure() to create a figure, set the title and labels with plt.title(), plt.xlabel(), and plt.ylabel(),

```python
#Code by Aditya kumar
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#importing the given archive Iris data file
df = pd.read_csv("Iris.csv")
```

```python
#Dataframe Head
df.head(10)
```

```python
#Line plotting
plt.figure(figsize=(8, 6))
plt.plot(df['SepalLengthCm'], df['SepalWidthCm'], marker='o', linestyle='-', color='g')
plt.xlabel("SepalLengthCm")
plt.ylabel("PetalLengthCm")
plt.title("IRIS DATA LINE PLOTING")
```

```python
#Scatter Plot
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'] ,label='Scatter Plot', color='red')
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("IRIS DATA LINE SCATTER PLOTING")
```

```python
#Scatter plot
plt.figure()
sp1 = df[df['Species']=='Iris-setosa']
sp2 = df[df['Species']=='Iris-virginica']
# Iris-setosa  vs  Iris-virginica
plt.scatter(sp1['PetalWidthCm'],sp2['PetalWidthCm'])
plt.xlabel("Iris-setosa PetalWidthCm")
plt.ylabel("Iris-virginica PetalWidthCm")
plt.title("Iris-setosa  vs  Iris-virginica")
```

```python
#Bar Plot
plt.bar(df['SepalLengthCm'], df['SepalWidthCm'], color='green')
plt.xlabel("SepalLengthCm")
plt.ylabel("SepalWidthCm")
plt.title("IRIS DATA BAR ploting using Matploy.pyplot ")
```

```python
#Box Plot
plt.boxplot(df['SepalLengthCm'], labels=['SepalLengthCm'])
```

```python
#Histrogram of Plot
plt.hist(df['SepalLengthCm'], label='Histogram', color='orange')
plt.title("Histogram Plot of Sepal Lenght")
```

```python
#Ploting using scripting layer of Histogram
plt.figure()
x = np.random.randn(100000)
plt.hist(x,100)
plt.title('Normal Distribution')
plt.show()
```

 Artist Layer Plotting

```python
# Create some sample data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
# Create a Figure and Axes object
fig, ax = plt.subplots(figsize=(8, 6))
# Create a Line2D object representing the sine curve
line, = ax.plot(x, y, label='Sine Curve', color='blue', linestyle='-', marker='o')
# Customize the plot by modifying properties of the Line2D object
line.set_linewidth(2)
line.set_markersize(8)
# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Artist Layer Plotting')
# Add a legend
ax.legend()
# Display the plot
plt.show()
```

```python
#Histrogram
ig = plt.figure()
ax2 = fig.add_axes([0.1, 0.3, 0.6, 0.7])
n, bins, patches = ax2.hist(np.random.randn(10000), 50,color='blue')
ax2.set_xlabel('X axis')
plt.show()
```

Scripting Layer

```python
#Ploting Bar chart
subjects = ['PPS', 'Maths', 'OS']
values = [90, 85, 95]
plt.bar(subjects, values, color='red', label='Marks')
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Marks Bar Chart')
plt.legend()
plt.show()
```

```python
#Horizonal bar
plt.barh(subjects, values,color="g")
plt.ylabel('Subjects')
plt.xlabel('Marks')
plt.title('Marks Bar Chart')
plt.legend()
plt.show()
```

```python
#Discrete Histrogram
data = np.random.randn(100)
plt.hist(data, bins=20, color='blue', edgecolor='black', alpha=0.7, label='Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.legend()
plt.show()
```
