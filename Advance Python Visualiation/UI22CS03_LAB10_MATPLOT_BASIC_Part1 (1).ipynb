# **Matplotlib Pyplot**

```python

import matplotlib.pyplot as plt
import numpy as np
```

### **1.1: Matplotlib Simple Line Ploting**
*Matplot is used to plot a simple line graphs using numpy generated array, in which we can specify the properties of line chart graph such as X axis labels, Y axis labels, Title at top of chart and legend etc.*


```python
#Created two distinct points on  x and y axis
x_axis= np.array([0,10])
y_axis = np.array([0,20])
print(x_axis,y_axis)

#Label the both axis
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")

#Ploting the both array using Plot with color and line attribute
plt.plot(x_axis,y_axis,'o:r')

#Assigned legend to it inside the chart at top right corner
plt.legend(["Simple Line Ploting"])

#Let's Show the Matplotted Chart Graph
plt.show()
```

```python
#Created one distinct points on y axis
y_points = np.array([0,3,7,2,4,10,8,11])
print(y_points)

#Label the both axis
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")

#Ploting the array using Plot with color and line attribute
plt.plot(y_points,'-.g',marker = 'o', ms = 10, mec = 'r')
plt.legend(["Simple Line Chart Ploting"])

#Writing the title at top of graph
plt.title("Dashed Pointd Formatted Line Graph")

#Embeeded the Grid Layout
plt.grid(axis='y')

#Let's Show the Matplotted Chart Graph
plt.show()
```

```python
#Created two distinct points on  x and y axis
y1= np.array([2,4,1,5,7,4])
y2= np.array([1,3,5,3,7,9])

#Label the both axis
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")

#Ploting the both array using Plot with color and line attribute
plt.plot(y1,'r')
plt.plot(y2,'g')
plt.title("2 Lines Representation",loc = 'center')

#Embeeded the Grid Layout
plt.grid()

#Let's Show the Matplotted Chart Graph
plt.show()
```

### **1.2: Display Multiple Plots :**
*The subplot() function takes three arguments that describes the layout of the figure.*

*The layout is organized in rows and columns, which are represented by the first and second argument.*

```python
#the figure has 1 row, 2 columns, and this plot is the first plot.
plt.subplot(1,2,1)
plt.plot(y_points,'g')
plt.grid()
plt.title("1st SubPlot")

#the figure has 1 row, 2 columns, and this plot is the second plot.
plt.subplot(1,2,2)
plt.plot(y1,'r')
plt.grid()
plt.title("2nd SubPlot")

#Let's Show the Matplotted Chart Graph
plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.suptitle("Multiple Sub-Plotings")

#Let's Show the Matplotted Chart Graph
plt.show()
```

### 1.3: **Matplotlib Scatter Plots** :
*With Pyplot, you can use the scatter() function to draw a scatter plot.*

*The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis.*

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)

#Let's Show the Matplotted Chart Graph
plt.show()
```

```python
x = np.array([2,5,7,3,2])
y = np.array([5,2,3,9,3])
colors= np.array(['red','green','blue','orange','hotpink'])
plt.scatter(x,y,c=colors)

#Let's Show the Matplotted Chart Graph
plt.show()
```

```python
#We can specify the colormap with the keyword argument cmap with the value of the colormap
#in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.
#In addition you have to create an array with values (from 0 to 100)

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.colorbar()

#Let's Show the Matplotted Chart Graph
plt.show()
```

```python
#Size
x = np.array([1,4,20,40,100])
y = np.array([1,3,6,40,80])
sizes = np.array([20,50,100,800,20])

plt.scatter(x, y, s=sizes)

#Let's Show the Matplotted Chart Graph
plt.show()
```

```python
#Combine Color Size and Alpha

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='nipy_spectral')

plt.colorbar()

#Let's Show the Matplotted Chart Graph
plt.show()
```

### 1.4: **Matplotlib Bars**
*With Pyplot, we can use the bar() function to draw bar graphs*

*Note: The categories and their values represented by the first and second argument as arrays.*

```python
x_bars= np.array(["Maths","OS","DBMS","PPS"])
y_marks= np.array([78,86,90,100])

plt.bar(x_bars,y_marks)
plt.title("Aditya Kumar EndSem Marks")
plt.xlabel("SUBJECTS")
plt.ylabel("MARKS SCORED")

#Let's Show the Matplotted Chart Graph
plt.show()
```

```python
plt.barh(x_bars,y_marks,color = "green",height=0.3)
plt.title("Aditya Kumar EndSem Marks")
plt.xlabel("MARKS SCORED")
plt.ylabel("SUBJECTS")

#Let's Show the Matplotted Chart Graph
plt.show()
```

### 1.5: **Matplotlib Histograms**
*A histogram is a graph showing frequency distributions.*

*It is a graph showing the number of observations within each given interval.*

```python
histrogram = np.random.normal(17, 10, 250)

plt.hist(histrogram)
plt.title("Simple Histrogram")
plt.xlabel("Number of Students")
plt.ylabel("Marks Scored")

#Let's Show the Matplotted Chart Graph
plt.show()
```

```python
 import random
 plt.figure()
 x = np.random.randn(10000)
 plt.hist(x,10000)
 plt.title("Normally Distributed 10000 random points Histrogram")
 plt.xlabel("Z scores")
 plt.ylabel("T Scores")

 #Let's Show the Matplotted Chart Graph
 plt.show
```

### 1.6: **Pie Charts**
With Pyplot, we can use the pie() function to draw pie charts, which is very
effetive way of data representation by visuals

```python
circle = np.array([10,20,30,40])
#10+20+30+40= 100, here basically 100% is a complete circle
mylabels = ["Maths", "OS", "DBMS", "PPS"]

plt.pie(circle,labels=mylabels)

#Let's Show the Matplotted Chart Graph

plt.show
```
