Seaborn is a advance statistical data visualization library built on top of Matplotlib in Python. To provides a high-level interface for creating attractive and informative statistical graphics.

```python
#Importing all the required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Here i am imporitng the data provided on google classroom from archive zip for data visualisation
iris = pd.read_csv("Iris.csv")
iris.head()
```

```python
#Priting the head of imported data
iris.head()
```

```python
#Seaplot Advance Line Chart Ploting
sns.lineplot(x="PetalLengthCm",y="PetalWidthCm",data=iris)
```

```python
#SNS Scatter Plot Based on Species of Petal
sns.scatterplot(x="PetalLengthCm",y="PetalWidthCm",data=iris,hue="Species")
```

```python
#Multiple BOX Ploting
del iris['Id'] #here Id was creating problem so deleted that dete from dataframe
sns.boxplot(data=iris)
```

```python
#Kernal Density Estimation (KDE PLOT)
sns.distplot(iris['SepalLengthCm'])
```

```python
#Advance Histrogram
sns.distplot(iris["PetalLengthCm"])
```

```python
#Correlation PLOT (HEATMAP)
df = iris.corr()
df
```

```python
#Create Advance Heat Map
sns.heatmap(df)
```

```python
#Linear Regression on Iris Data
sns.lmplot(x="PetalWidthCm",y="PetalLengthCm",data=iris)
```

```python
#method 2 to data visualize of linear regresion by seaborn advance data analysis
sns.regplot(x="PetalWidthCm",y="PetalLengthCm",data=iris)
```
