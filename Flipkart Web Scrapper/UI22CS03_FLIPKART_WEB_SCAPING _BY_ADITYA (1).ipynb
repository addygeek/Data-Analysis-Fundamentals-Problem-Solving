```python
#Importing all requiued Modules
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
#Enter here the name of product you want to search 
search_product= input("Enter the Product Name:  ")
#Now i have used the qurery language to search and make dyanimc url of flipkart to get the keyword search
flipkart_link="https://www.flipkart.com/search?q="+search_product+"&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
print(flipkart_link)
```

```python
#Let's Now Request the content of Flipkart webpage 
url_req=requests.get(flipkart_link)
```

```python
# soup is varible which is using BeautifulSoup Module to parse the html contents of webpage
soup = bs(url_req.content, 'html.parser')
```

```python
#Let's Print out the webpage 
#Note:- Use the stable internet connection to get it earlier i getting error 500 because of network multiple requests issue
print(soup.prettify())
```

```python
#Let's Now display the names of all product name and store all in a list
#Here from DOM of webpage, I have identify the class name which is repeating on every product name
Product_name = soup.find_all("div",class_="_4rR01T")
print(Product_name)
#Created List in name of Products, Rating which will store the below data in it 
products=[]
ratings=[]
```

```python
#Let's Now display the prices of all product and store all in a list
#Here from DOM of webpage, I have identify the class name which is repeating on every product price tag
Product_price=  soup.find_all("div",class_="_30jeq3 _1_WHN1")
print(Product_price)
#Created a list prices where all the prices will get stored, later in next cell will append it all at once
prices=[]
```

```python
#Here we are appending all the attributes like Name, Price and Raiting 
for a in soup.findAll(attrs={'class':'_2kHMtA'}):
    name=a.find('div', attrs={'class':'_4rR01T'})
    price=a.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
    rating=a.find('div', attrs={'class':'_3LWZlK'})
    products.append(name.text)
    prices.append(price.text)
    ratings.append(rating.text)
```

```python
# Let's Print it out all list of Product Name
print(products)
```

```python
# Let's Print it out all list of Product Prices
print(prices)
```

```python
# Let's Print it out all list of Product Raiting
print(ratings)
```

```python
#Let's Display the all Product name, Product Price and Product Rating in a single row 
for i in range(len(products)):
    print(products[i] + ' avaiable at: ' + prices[i]+ ' with '+ratings[i]+' rating')
```

```python
#Let's display it more better using columns name and store in a 2D array
product_details=pd.DataFrame(list(zip(products, prices, ratings)),columns=['Product Name','Price', 'Ratings'])
```

```python
#Let's Print it out all data in one frame
print(product_details)
```

```python
#let's convert the above in a Dataframe so can perform multiple data analysis
df= pd.DataFrame(product_details)
```

```python
#Let's Display the Prices in sorted form (i.e Show me the chepest phone first)
df.sort_values(by="Price", ascending=True)
```
